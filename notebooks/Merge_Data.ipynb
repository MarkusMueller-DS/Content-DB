## Unify columns

- with different sources i get various data
- folloing are the essential columns
    - title
    - summary / first paragraph
    - ulr
    - source

```python
# add source to KDnuggets
KD_source_list = ['KDnuggets'] * df_kdnuggets.shape[0]
df_kdnuggets['source'] = KD_source_list 

# add source to pocket
PO_source_list = ['Pocket'] * df_pocket.shape[0]
df_pocket['source'] = PO_source_list
```

```python
# rename columns
df_dataelixir =  df_dataelixir.rename(columns={"link": "url"})

#rename content to summary pocket
df_pocket = df_pocket.rename(columns={'content': 'summary'})
```

```python
df_list = [df_pocket, df_dataelixir, df_kdnuggets]

df_combined = pd.concat(df_list)
```

this is a mess i need to delete some columns

```python
df_combined = df_combined.drop(['Unnamed: 0', 'item_id', 'date', 'issue', 'content', 'tag'], axis=1)
```

```python
cols = ['title', 'summary', 'url', 'source']
df_combined = df_combined[cols]
```
