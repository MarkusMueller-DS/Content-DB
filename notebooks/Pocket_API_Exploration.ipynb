```python
# imports
import json 
import pandas as pd
import requests
import sys
sys.path.append('/Users/markusmuller/python/projects/content-db/pocket')
import pocket_api_tokens
import pickle
```

```python
con_key = pocket_api_tokens.consumer_key
access_token = pocket_api_tokens.access_token

sep = '&'
stripped_token = access_token.split(sep, 1)[0]
```

form the pocket api: Whenever possible, you should use the since parameter, or count and and offset parameters when retrieving a user's list. After retrieving the list, you should store the current time (which is provided along with the list response) and pass that in the next request for the list. This way the server only needs to return a small set (changes since that time) instead of the user's entire list every time.

```python
# https://getpocket.com/developer/docs/v3/retrieve
# define paramters for the query
parameters = {"consumer_key":con_key, "access_token":stripped_token, "count":"1"}
```

```python
# use parameters to get the data
r = requests.get("https://getpocket.com/v3/get", params=parameters)
```

```python
r.json()
```

### Parse Json to get the data

```python
data = request_obj.json()
```

```python
data.keys()
```

```python
data['list'].keys()
```

```python
# number of saved items
len(data['list'].keys())
```

```python
# with get and the key() from list i can get the individual data from each saved item
data['list'].get('2425410382')
```

Which values are intersting for me:
- item_id
- resolved_url (if the given url was shortend)
- time_added (which is encoded as a timestamp)
- resolved_title
- excerpt (first few lines of the item)

```python
item_id = []
url = []
date = []
title = []
content = [] 
```

```python
for id_ in data['list'].keys():
    item_id.append(id_)
    url.append(data['list'].get(id_)['resolved_url'])
    date.append(data['list'].get(id_)['time_added'])
    title.append(data['list'].get(id_)['resolved_title'])
    content.append(data['list'].get(id_)['excerpt'])
```

```python
df = pd.DataFrame()
df['item_id'] = item_id
df['url'] = url
df['date'] = date
df['title'] = title
df['content'] = content
```
